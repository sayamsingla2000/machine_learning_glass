```python
a = ['Ag','Al','Am','Ar','As','Au','B','Ba','Be','Bi','Br','C','Ca','Cd','Ce','Cf','Cl','Co','Cr','Cs','Cu','Dy','Er','Eu','F','Fe','Ga','Gd','Ge','H','He','Hf','Hg','Ho','I','In','Ir','K','La','Li','Lu','Mg','Mn','Mo','N','Na','Nb','Nd','Ne','Ni','Np','O','Os','P','Pb','Pd','Pr','Pt','Ra','Rb','Re','Rh','Ru','S','Sb','Sc','Se','Si','Sm','Sn','Sr','Ta','Tb','Te','Th','Ti','Tl','Tm','U','V','W','Xe','Y','Yb','Zn','Zr']
a10 = ['Al','As','B','Ba','Bi','Cd','Cu','Dy','Er','Fe','Ga','Gd','Ge','In','K','La','Mn','Na','P','Pb','Pr','S','Sb','Se','Si','Sn','Tb','Te','Tl','Xe','Zn']
```

```python
len(a)
```

```python
def reject_outliers(data, m = 2.):
        d = np.abs(data - np.median(data))
        mdev = np.median(d)
        s = d/mdev if mdev else 0.
        return (s < m)

def mean_dup(x_):
        global reject_outliers
        if 1==len(np.unique(x_.values)):
            return x_.values[0]
        else:
            x = x_.values[reject_outliers(x_.values.copy())]
            x_mean = x.mean()
            mask = (x_mean*0.995 <= x) & (x <= x_mean*1.005)
            return x[mask].mean()
```

```python
df_T_v = df_T_v[df_T_v[a].sum(axis=1).between(99,101)]
#md_columns[:-1] taking only those features on the basis of which we want to remove duplicates. here it is same as md_features.
df_T_v = df_T_v.groupby(a,as_index=False).agg(mean_dup)
df_T_v = df_T_v.dropna()
df =df_T_v
```
